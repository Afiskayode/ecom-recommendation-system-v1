## Data Preparation & Feature Transformation

```python
import pandas as pd
import os

data_path = "../data/"
```

### Creating clean customers to product interaction dataset

```python
customers = pd.read_csv(os.path.join(data_path, 'olist_customers_dataset.csv'))
order_items = pd.read_csv(os.path.join(data_path, 'olist_order_items_dataset.csv'))
order_reviews = pd.read_csv(os.path.join(data_path, 'olist_order_reviews_dataset.csv'))
products = pd.read_csv(os.path.join(data_path, 'olist_products_dataset.csv'))
orders = pd.read_csv(os.path.join(data_path, 'olist_orders_dataset.csv'))

merged_data = customers.merge(orders, on='customer_id')
merged_data = merged_data.merge(order_items, on='order_id')
merged_data = merged_data.merge(order_reviews, on='order_id')
merged_data = merged_data.merge(products, on='product_id')

df = merged_data[['customer_unique_id', 'product_id', 'review_score']]
```

#### Filling reviews score with 3 where a customer purchased an item but didnt leave a review; we are assuming since they did leave bad review they liked the product 

```python
df['review_score'] = df['review_score'].fillna(3)
```

### Taking product with interaction greater than 10 and saving the new dataset

```python
minimum_product_interaction = 10
product_count = df['product_id'].value_counts()
popular_product = product_count[product_count > minimum_product_interaction].index

df_clean = df[df['product_id'].isin(popular_product)].copy()

df_clean.to_csv((os.path.join(data_path, "clean_interactions.csv")), index=False)
```
